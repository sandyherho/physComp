```python
def dotProduct(a,b):
    c = a[0] * b[0] + a[1] * b[1] + a[2] * b[2]
    return c
```

```python
x3, y3 = np.meshgrid(np.linspace(-10,10,201), np.linspace(-10,10,201))
z3 = np.cos(x3 + y3) + 0.05*(x3 - y3)

plt.contourf(x3,y3,z3);
plt.colorbar();
```

```python
plt.figure(figsize=(15,8))
proj3d = plt.axes(projection='3d')
proj3d.contour3D(x3,y3,z3,100, cmap='binary');
proj3d.view_init(70,30);
```

```python
plt.figure(figsize=(15,8));
plt3D = plt.axes(projection='3d');

z1 = np.linspace(0,30,301)
x1 = np.sin(z1)
y1 = np.cos(z1)

plt3D.plot3D(x1, y1, z1);
plt3D.set_xlabel('$x$');
plt3D.set_ylabel('$y$');
plt3D.set_zlabel('$z$');
```
